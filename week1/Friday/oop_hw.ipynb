```python
class ComputerComponent:
    def __init__(self,manufacturer,model,serial_number,name):
        self.__manufacturer = manufacturer
        self.__model = model
        self.__serial_number = serial_number
        self.__name = name


    def display_info(self):
        print(f"name: {self.__name}")
        print(f"\tmanufacturer: {self.__manufacturer}")
        print(f"\tmodel: {self.__model}")
        print(f"\tserial number: {self.__serial_number}")

    # Getter methods
    def get_manufacturer(self):
        return self.__manufacturer

    def get_model(self):
        return self.__model

    def get_serial_number(self):
        return self.__serial_number

    def get_name(self):
        return self.__name

    # Setter methods
    def set_manufacturer(self, manufacturer):
        self.__manufacturer = manufacturer

    def set_model(self, model):
        self.__model = model

    def set_serial_number(self, serial_number):
        self.__serial_number = serial_number

    def set_name(self, name):
        self.__name = name
```

```python
class CPU(ComputerComponent):
    def __init__(self,manufacturer,model,serial_number,name,cores,clock_speed,cache_size):
        self.__cores = cores
        self.__clock_speed = clock_speed
        self.__cache_size = cache_size
        super().__init__(manufacturer,model,serial_number,name)

    def display_info(self):
        super().display_info()
        print(f"\tcores: {self.__cores}")
        print(f"\tclock speed: {self.__clock_speed}")
        print(f"\tcache size: {self.__cache_size}")

    def get_cores(self):
        return self.__cores

    def get_clock_speed(self):
        return self.__clock_speed

    def get_cache_size(self):
        return self.__cache_size

    def set_cores(self, cores):
        self.__cores = cores

    def set_clock_speed(self, clock_speed):
        self.__clock_speed = clock_speed

    def set_cache_size(self, cache_size):
        self.__cache_size = cache_size
```

```python

class Memory(ComputerComponent):
    def __init__(self, manufacturer, model, serial_number, name, capacity):
        self.__capacity = capacity
        super().__init__(manufacturer, model, serial_number, name)


    def display_info(self):
        super().display_info()
        print(f"\tcapacity: {self.__capacity}")

    def get_capacity(self):
        return self.__capacity

    def set_capacity(self, capacity):
        self.__capacity = capacity
```

```python

class Storage(ComputerComponent):
    def __init__(self,manufacturer,model,serial_number,name,storage_type,size):
        self.__storage_type = storage_type
        self.__size = size
        super().__init__(manufacturer,model,serial_number,name)

        
    def display_info(self):
        super().display_info()
        print(f"\tstorage type: {self.__storage_type}")
        print(f"\tsize: {self.__size}")

    def get_storage_type(self):
        return self.__storage_type

    def get_size(self):
        return self.__size

    def set_storage_type(self, storage_type):
        self.__storage_type = storage_type

    def set_size(self, size):
        self.__size = size
```

```python
class Computer:
    def __init__(self,cpu,memory,storage,operating_system):
        self.__check_for_comp_components(cpu,memory,storage) 
        self.__operating_system = operating_system
        self.__cpu = cpu
        self.__memory = memory
        self.__storage = storage

    def display_info(self):
        print(f"operating system {self.__operating_system}")
        self.__cpu.display_info()
        self.__memory.display_info()
        self.__storage.display_info()
    def __check_for_comp_components(self,*args):
        for comp in args:
            if isinstance(comp,ComputerComponent) == False:
                raise TypeError("A computer can only take computer components")
            
    def replace_component(self,component:ComputerComponent):
        self.__check_for_comp_components(component)
        if component.get_name() == "cpu":
            self.__cpu = component
        elif component.get_name() == "memory":
            self.__memory = component
        elif component.get_name() == "storage":
            self.__storage = component
```

```python
my_cpu = CPU("cpuM","cpuMo","cpu_serial","cpu",4,5.6,50)
my_memory = Memory("memoryM","memoryMo","memory_serial","memory",40)
my_storage = Storage("storageM","storageMo","storage_serial","storage","gigs",60)
my_computer = Computer(my_cpu, my_memory, my_storage, "Windows 10")
my_computer.display_info()
my_computer.replace_component(my_memory)
```
